```python
import cv2
```

```python
img=cv2.imread('/content/Madhuri image (1)')
```

```python
import time
start=time.time()
```

```python
# Internal mechanism
```

```python
grey=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
from google.colab.patches import cv2_imshow
cv2_imshow(grey)
```

```python
invert=cv2.bitwise_not(grey)
cv2_imshow(invert)
```

```python
blur=cv2.GaussianBlur(invert,(15,15),0)
cv2_imshow(blur)
```

```python
invertblur=cv2.bitwise_not(blur)
cv2_imshow(invertblur)
```

```python
sketch=cv2.divide(grey,invertblur,scale=250.0)
cv2_imshow(sketch)
```

```python
from  google.colab.patches import cv2_imshow
cv2.imwrite('Madhuri.png',sketch)
```
